```python
def get_weight_class_correlations(df, weight_class_name):
    """
    Returns the top 10 and bottom 10 correlations between fight stats and winning,
    for a specific weight class.
    """
    # Filter for one weight class
    wc_df = df[df['weight_class'] == weight_class_name]
    
    # Only numeric columns (excluding 'won' for now)
    numeric_cols = wc_df.select_dtypes(include='number').columns.drop('won', errors='ignore')
    
    # Prepare a DataFrame for correlation
    corr_df = wc_df[numeric_cols.tolist() + ['won']]
    
    # Calculate correlation with 'won'
    correlations = corr_df.corr()['won'].dropna()
    
    # Get top 10 and bottom 10
    top_10 = correlations.sort_values(ascending=False).head(10)
    bottom_10 = correlations.sort_values(ascending=True).head(10)
    
    return top_10, bottom_10
```

```python
def get_all_correlations_by_weight(df, target_col='won'):
    weight_classes = df['weight_class'].dropna().unique()
    
    # Define non-combat columns
    noncombat_cols = [
        'age', 'Height_cms', 'Reach_cms', 'Weight_lbs',
        'wins', 'losses', 'draw',
        'current_win_streak', 'current_lose_streak',
        'longest_win_streak', 'total_rounds_fought',
        'total_time_fought(seconds)', 'total_title_bouts'
    ]
    
    # Define differential columns
    diff_cols = ['reach_diff', 'height_diff', 'win_streak_diff', 'age_diff']
    
    results = {}

    for wc in weight_classes:
        subset = df[df['weight_class'] == wc]

        if len(subset) < 100:
            continue

        result = {}

        # Overall top/bottom 10 numeric correlations
        num_cols = subset.select_dtypes(include='number').columns
        corr_all = subset[num_cols].corr()[target_col].dropna()
        result['Top 10 Overall'] = corr_all.sort_values(ascending=False).head(10)
        result['Bottom 10 Overall'] = corr_all.sort_values().head(10)

        # Non-combat only
        corr_noncombat = subset[noncombat_cols + [target_col]].corr()[target_col].dropna()
        result['Non-Combat'] = corr_noncombat.sort_values(ascending=False)

        # Differential only
        corr_diff = subset[diff_cols + [target_col]].corr()[target_col].dropna()
        result['Differentials'] = corr_diff.sort_values(ascending=False)

        results[wc] = result

    return results

# Run the master function
full_corr_by_weight = get_all_correlations_by_weight(df_fighter_stats)
```

```python
def get_all_correlations_by_weight(df, target_col='won'):
    results = {}
    
    weight_classes = df['weight_class'].dropna().unique()
    
    # Define column groups
    non_combat_cols = [
        'current_win_streak', 'current_lose_streak', 'longest_win_streak',
        'wins', 'losses', 'draw', 'total_time_fought(seconds)',
        'total_rounds_fought', 'total_title_bouts', 'Reach_cms', 'Height_cms',
        'Weight_lbs', 'age'
    ]
    
    differential_cols = ['reach_diff', 'height_diff', 'win_streak_diff', 'age_diff']
    
    for wc in weight_classes:
        subset = df[df['weight_class'] == wc].copy()
        
        # Skip if not enough data
        if subset.shape[0] < 30:
            continue
        
        result = {}

        # Overall correlations
        num_cols = subset.select_dtypes(include='number').columns
        if target_col not in num_cols:
            continue
        
        corr_all = subset[num_cols].corr()[target_col].dropna()
        result['Top 10 Overall'] = corr_all.abs().sort_values(ascending=False).head(10)
        result['Bottom 10 Overall'] = corr_all.abs().sort_values().head(10)

        # Non-combat correlations
        non_combat_present = [col for col in non_combat_cols if col in subset.columns]
        if non_combat_present:
            result['Non-Combat'] = subset[non_combat_present + [target_col]].corr()[target_col].dropna()

        # Differential correlations
        diff_present = [col for col in differential_cols if col in subset.columns]
        if diff_present:
            result['Differentials'] = subset[diff_present + [target_col]].corr()[target_col].dropna()

        results[wc] = result

    return results
```

```python
def get_cleaned_correlations_by_weight(df):
    results = {}
    weight_classes = df['weight_class'].dropna().unique()

    for wc in weight_classes:
        subset = df[df['weight_class'] == wc]

        # Skip if not enough data
        if subset[target_col].nunique() < 2 or len(subset) < 50:
            continue

        # --- Overall Top & Bottom 10 Correlations (absolute, excluding 'won' itself)
        numeric_cols = subset.select_dtypes(include='number').columns.drop(target_col)
        corr_series = subset[numeric_cols].corrwith(subset[target_col]).dropna()
        top10 = corr_series.abs().sort_values(ascending=False).head(10)
        bottom10 = corr_series.abs().sort_values(ascending=True).head(10)

        # --- Non-Combat Correlations (sorted by actual correlation, not abs)
        non_combat_clean = [
            col for col in non_combat_cols if col in subset.columns
        ]
        nc_corrs = subset[non_combat_clean + [target_col]].corr()[target_col].drop(target_col)
        nc_corrs = nc_corrs.sort_values(ascending=False)

        # --- Differential Correlations
        diff_clean = [col for col in diff_cols if col in subset.columns]
        diff_corrs = subset[diff_clean + [target_col]].corr()[target_col].drop(target_col)
        diff_corrs = diff_corrs.sort_values(ascending=False)

        # --- Stance win rates
        if 'Stance' in subset.columns:
            stance_win_rates = subset.groupby('Stance')[target_col].mean().sort_values(ascending=False)
            stance_counts = subset['Stance'].value_counts()
            stance_data = pd.DataFrame({
                'win_rate': stance_win_rates,
                'count': stance_counts
            }).dropna()
        else:
            stance_data = pd.DataFrame()

        results[wc] = {
            'Top 10 Absolute Correlations': corr_series[top10.index],
            'Bottom 10 Absolute Correlations': corr_series[bottom10.index],
            'Non-Combat Correlations': nc_corrs,
            'Differential Correlations': diff_corrs,
            'Stance Win Rates': stance_data
        }

    return results
```

```python
def get_cleaned_correlation_summary_by_class(df, weight_class):
    """
    Returns a cleaned summary of correlation results for a specific weight class.
    Includes:
    - Top 10 Absolute Correlations
    - Bottom 10 Absolute Correlations
    - Non-Combat Stat Correlations
    - Differential Correlations
    - Stance Win Rates
    """
    subset = df[df['weight_class'] == weight_class].copy()

    # Ensure 'won' is numeric
    subset['won'] = subset['won'].astype(float)

    # === 1. Top/Bottom 10 Absolute Correlations ===
    numeric_cols = subset.select_dtypes(include='number').columns
    corr_all = subset[numeric_cols].corr()['won'].dropna()
    corr_all_abs = corr_all.drop(labels='won').abs().sort_values(ascending=False)
    top_10 = corr_all_abs.head(10).index
    bottom_10 = corr_all_abs.tail(10).index

    top_10_corr = corr_all[top_10]
    bottom_10_corr = corr_all[bottom_10]

    # === 2. Non-Combat Stats ===
    non_combat = [
        'current_win_streak', 'current_lose_streak', 'longest_win_streak',
        'wins', 'losses', 'draw', 'total_time_fought(seconds)', 'total_rounds_fought',
        'total_title_bouts', 'Reach_cms', 'Height_cms', 'Weight_lbs', 'age'
    ]
    non_combat_corr = subset[non_combat + ['won']].corr()['won'].drop('won').sort_values(ascending=False)

    # === 3. Differentials ===
    diffs = ['reach_diff', 'height_diff', 'win_streak_diff', 'age_diff']
    diff_corr = subset[diffs + ['won']].corr()['won'].drop('won').sort_values(ascending=False)

    # === 4. Stance win rates ===
    stance_counts = subset['Stance'].value_counts()
    stance_winrates = subset.groupby('Stance')['won'].mean()
    stance_summary = pd.DataFrame({'win_rate': stance_winrates, 'count': stance_counts})
    stance_summary = stance_summary[stance_summary['count'] >= 30]  # Filter rare stances

    return {
        'Top 10 Absolute Correlations': top_10_corr,
        'Bottom 10 Absolute Correlations': bottom_10_corr,
        'Non-Combat Correlations': non_combat_corr,
        'Differential Correlations': diff_corr,
        'Stance Win Rates': stance_summary
    }
```

```python
def get_correlation_summary_by_class(df, weight_class):
    subset = df[df['weight_class'] == weight_class].copy()

    # Clean missing values
    subset = subset.dropna(subset=['won'])

    # --- 1. Top 10 Correlations (closest to +1) ---
    num_cols = subset.select_dtypes(include='number').columns
    corr = subset[num_cols].corr()['won'].drop('won')

    top_10 = corr.sort_values(ascending=False).head(10)

    # --- 2. Bottom 10 Correlations (closest to -1) ---
    bottom_10 = corr.sort_values(ascending=True).head(10)

    # --- 3. Non-Combat Correlations (ordered from most to least positive) ---
    non_combat = [
        'current_win_streak', 'current_lose_streak', 'longest_win_streak',
        'wins', 'losses', 'total_time_fought(seconds)', 'total_rounds_fought',
        'total_title_bouts', 'Reach_cms', 'Height_cms', 'Weight_lbs', 'age', 'draw'
    ]
    non_combat_corr = subset[non_combat + ['won']].corr()['won'].drop('won').sort_values(ascending=False)

    # --- 4. Differential Correlations (diff stats) ---
    diff_cols = ['reach_diff', 'height_diff', 'win_streak_diff', 'age_diff']
    diff_corr = (
        subset[diff_cols + ['won']]
        .corr()['won']
        .drop('won')
        .sort_values(ascending=False)
    )

    # --- 5. Stance Win Rates ---
    stance_data = (
        subset.groupby('Stance')['won']
        .agg(['mean', 'count'])
        .rename(columns={'mean': 'win_rate'})
        .sort_values(by='count', ascending=False)
        .head(5)
    )

    return {
        'Top 10 Correlations': top_10,
        'Bottom 10 Correlations': bottom_10,
        'Non-Combat Correlations': non_combat_corr,
        'Differential Correlations': diff_corr,
        'Stance Win Rates': stance_data
    }

# Example usage:
lightweight_corrs = get_correlation_summary_by_class(df_fighter_stats, 'Lightweight')
```

```python
def rename_series(index, name_map):
    return [name_map.get(name, name) for name in index]
```
